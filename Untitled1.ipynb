```python
# assuming the x and y coordinates if of the middle of the frame
num_pixel_x = 2304
num_pixel_y = 2304

if num_pixel_x === num_pixel_y:
    centering_factor = num_pixel_x/2 

pixel_in_microns = 0.65

# the x and y gloabl coordinates of the center of the frame in microns
dXpos = -33208.9
dYpos = 3523.8
```

```python
def convertToGlobal(pix_x, pix_y, centering_factor,dXpos, dYpos):
    global_x = dXpos/centering_factor * pix_x
    global_y = dYpos/centering_factor * pix_y
    return(global_x, global_y)
```

```python
new_x, new_y = convertToGlobal(1147, 1172, centering_factor, dXpos, dYpos)
```

```python
zoomed_out = opencv.imread('.\Frozen_4x.tif')
```

```python
plt.imshow(zoomed_out)
```

```python
def convertToPixel(glob_x, glob_y, centering_factor, dXpos, dYpos):
    pix_x = centering_factor/dXpos * glob_x
    pix_y = centering_factor/dYpos * glob_y
    return(pix_x, pix_y)
```

```python
zoom_out_dXpos = -33208.0
zoom_out_dYpos = 3523.8
zoom_out_centering_factor = centering_factor
```

```python
zoomOut_pix_x, zoomOut_pix_y = convertToPixel(new_x, new_y, 
                                              zoom_out_centering_factor, zoom_out_dXpos, zoom_out_dYpos)
```

```python
zoomed_out = opencv.circle(zoomed_out, (int(zoomOut_pix_x), int(zoomOut_pix_y)), 20, (255, 255, 0), -1)
plt.imshow(zoomed_out)
```
